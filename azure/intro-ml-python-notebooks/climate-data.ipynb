# Climate Change Analysis

```python
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
sns.set()
```

```python
yearsBase, meanBase = np.loadtxt('data/5-year-mean-1951-1980.csv', delimiter=',', usecols=(0, 1), unpack=True)
years, mean = np.loadtxt('data/5-year-mean-1882-2014.csv', delimiter=',', usecols=(0, 1), unpack=True)
```

```python
plt.scatter(yearsBase, meanBase)
plt.title('Scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()
```

# Linear Regression with Numpy

```python
# Least squares polynomial fit
m,b = np.polyfit(yearsBase, meanBase, 1)

# lets create a y=mx+b function
def f(x):
    return m*x + b

plt.scatter(yearsBase, meanBase)
plt.plot(yearsBase, f(yearsBase))
plt.title('scatter plot of mean temp difference vs years')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()

print(f'y = {m}*x + {b}')
plt.show()
```

Linear Regression with Scikit Learn

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(fit_intercept=True)

regressor.fit(yearsBase[:, np.newaxis], meanBase)
preds = regressor.predict(yearsBase[:, np.newaxis])

plt.scatter(yearsBase, meanBase)
plt.plot(yearsBase, preds)
plt.title('Scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()

print(f'y = {regressor.coef_[0]}*x + {regressor.intercept_}')
```
